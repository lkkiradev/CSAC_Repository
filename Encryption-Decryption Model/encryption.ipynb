This model leverages a combination of split and process technique (ensembling divide and conquer) and manipulation of ASCII values to enhance data security. By splitting the input data into smaller segments, this approach adds layers of complexity to the encryption process.

Central to this model is the manipulation of ASCII values, which allows for a dynamic transformation of the original data. By applying mathematical operations to the ASCII values of characters, we create a unique mapping that jumbles the plaintext, making it resistant to common cryptographic attacks. The purpose of this model is that even if an attacker gains access to a portion of the encrypted data, deciphering the entire message should remain quite a challenge.

The scope of this model accepts user input only within the range of characters with ASCII value of 32 to 126, with a maximum of 20 characters per input. The first 32 characters in the ASCII-table are unprintable control codes and are used to control peripherals such as printers. Codes 32-127 are common for all the different variations of the ASCII table, they are called printable characters, represent letters, digits, punctuation marks, and a few miscellaneous symbols. However, character 127 represents the command DEL.

Now, we split the overall length of the message and apply different formula or calculation for each halves.

```python
df['1st_half'] = df['0'] + df['1']
df['2nd_half'] = df['2'] + df['3']
df['ASCII[1st_half]'] = df['ASCII[0]'].astype(str) + ', ' + df['ASCII[1]'].astype(str)
df['ASCII[2nd_half]'] = df['ASCII[2]'].astype(str) + ', ' + df['ASCII[3]'].astype(str)
```

Now that we split the characters and their ascii values and grouped them into halves, we can now start our calculation. In our model, each half has different arithmetic formula or treatment for their shifts. We also used a number randomizer ranging from 1 to 32, that will be used in creating our shift values. This creates a total number of 32 possible encryptions for each input.



```python
def randomize_number(min,max):
    return r.randint(min,max)

min_base = 1
max_base = 32
random_number = randomize_number(min_base,max_base)
random_number
```

Now that we've obtained a random number, we can now formulate our shift value. Below is the code for the shift_value:

```python
# where n is the length of input; 
# and midpoint is the middle number after splitting the characters for each half of the string:

print("Random number: " , random_number)
def encrypt_half(message, is_first_half):
    n = len(message)
    shift_keys=[]
    shift_keys2=[]

    for i in range(n):
       
        shift_value = random_number + i 
        if shift_value % 2 == 0: 
            if is_first_half:
                final_shift = shift_value + (i + (i ** 2))
                shift_keys.append(final_shift)
            else:
                final_shift = shift_value + 2 * (i + (i ** 2))
                shift_keys2.append(final_shift)
        else: 
            if is_first_half:
                final_shift = shift_value + (i + (i ** 3))
                shift_keys.append(final_shift)
            else:
                final_shift = shift_value + 2 * (i + (i ** 3))
                shift_keys2.append(final_shift)

        print(f"[Encrypt] Char: {message[i]}, Final Shift: {final_shift}")
        column = 'Final Shift[first_half]' if is_first_half else 'Final Shift[2nd_half]'
        df.at[0, column] = str(shift_keys) if is_first_half else str(shift_keys2)
        
sample_message = df['Message'].iloc[0]
midpoint = len(sample_message) // 2
first_half_of_the_input = sample_message[:midpoint]
second_half_of_the_input = sample_message[midpoint:]

encrypt_half(first_half_of_the_input,is_first_half=True)
encrypt_half(second_half_of_the_input,is_first_half=False)
```


### Shift Value Formula

The shift value contains the sum of random number generated and the current index of the character of a half.

### shift_value = random_number + i 

sample message: fish

random_number = 32

        for the first half of the message:

####        f 0 = 32 + 0 = 32

####        i 1 = 32 + 1 = 33 

        for the second half of the message

####        s 0 = 32 + 0 = 32

####        h 1 = 32 + 1 = 33 



### Final Shift Value Formula

The final shift value contains the sum of the shift value and the square or cube of the index added by itself then multiplied by one or two, depending on which half the character is in and if its obtained shift value is odd or even:

### if it is the first half, then:

            if shift_value is even, then: final_shift_value = shift_value + (i + (i^2))
    
            if shift_value is odd, then: final_shift_value = shift_value + (i + (i^3))

### if it is the second half, then:

            if shift_value is even, then: final_shift_value = shift_value + 2*(i + (i^2))

            if shift_value is odd, then: final_shift_value = shift_value + 2*(i + (i^3))


sample message: fish

### if it is the first half, then:

        if shift_value is even, then: 

            f [0] = 32 + (0 + 0^2) = 32
    
        if shift_value is odd, then:

            i [1] = 33 + (1 + 1^3) = 35

### if it is the second half, then:

        if shift_value is even, then: 

            s [0] = 32 + 2(0 + 0^2) = 32
    
        if shift_value is odd, then:
    
            h [1] = 33 + 2(1 + 1^3) = 37


#### Note:
Since the computation for the shift value is the sum of random_number and current index, 
the shift value will always be even, if:
 
        the character's current index == even and random_number == even
        the character's current index == odd and random_number == odd

After formulating each shift value for every characters in an input, we can now calculate for the final ASCII value for our encrypted keys and ciphered text. The calculation is as follows:

#### ascii_value = (ord(sample_message[i]) + final_shift) - round(m.log(final_shift))

where: 
- (ord(sample_message[i]) is a py func that gets the ASCII value of each character in the message based on the ASCII table;
- final_shift is the shift we calculated earlier based on certain conditions (is_first_half, is_even); and
- round(m.log(final_shift) from the math package, which obtains only the rounded whole number of the natural logarithm of the acquired final_shift value


### For encryption calculation:

example:

ascii_value = (ord(sample_message[i]) + final_shift) - round(m.log(final_shift)

sample message: fish

f = ascii_value = (102 + 32) - int(log(32)) = 134 - 3 = 131

#### while ascii_value > 126:
            ascii_value = 31 + (ascii_value - 126)

#### then:
            encrypted_key = 31 + (131 - 126) = 36            

for the character 'f' in the message 'fish', the calculated encrypted key of our model is 36.

### Reverse arithmetic of encryption (decryption calculation):

encrypted_key for 'f' =  36.

Input : f

Ciphered text : $

decrypted_value = (encrypted_keys[i] - final_shift) + round(m.log(final_shift))

decrypted_value = (36 - 32) + 3 = 7 

    while ascii_value < 32:
        ascii_value = 126 - (31 - decrypted_value)
        decrypted_value = 126 - (31 - 7) = 102

decrypted_key = 102

equivalent character for 102 = 'f'


```python
# where n is the length of input; 
# and midpoint is the middle number after splitting the characters for each half of the string:

print("Random number: " , random_number)
def encrypt_half(message, is_first_half):
    n = len(message)
    ek_1=[]
    ek_2=[]
    encrypted_keys=[]

    for i in range(n):
       
        shift_value = random_number + i 
        if shift_value % 2 == 0: 
            if is_first_half:
                final_shift = shift_value + (i + (i ** 2))
                
            else:
                final_shift = shift_value + 2 * (i + (i ** 2))
                
        else: 
            if is_first_half:
                final_shift = shift_value + (i + (i ** 3))
                
            else:
                final_shift = shift_value + 2 * (i + (i ** 3))
                
        
        ascii_value = (ord(message[i]) + final_shift) - round(m.log(final_shift))
       
        while ascii_value > 126:
            ascii_value = 31 + (ascii_value - 126)
        
        ek_1.append(ascii_value) if is_first_half else ek_2.append(ascii_value)
        column = 'EncryptedKeys[1st]' if is_first_half else 'EncryptedKeys[2nd]'
        df.at[0, column] = str(ek_1) if is_first_half else str(ek_2)
        
    encrypted_keys = ek_1 + ek_2
    ciphered_text = ''.join(chr(key) for key in encrypted_keys)
    print("encrypted keys: ", encrypted_keys)
    print("Ciphered text: ", ciphered_text)    
    
sample_message = df['Message'].iloc[0]
midpoint = len(sample_message) // 2
first_half_of_the_input = sample_message[:midpoint]
second_half_of_the_input = sample_message[midpoint:]

encrypt_half(first_half_of_the_input,is_first_half=True)
encrypt_half(second_half_of_the_input,is_first_half=False)
```
